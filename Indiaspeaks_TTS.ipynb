```python
from TTS.api import TTS
import pandas as pd
import os

# Load your dataset
df = pd.read_csv('/content/toy_tts_parallel_data.csv')

# Load model
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False, gpu=False)

# Create output folder
os.makedirs("output_wavs", exist_ok=True)

# Generate audio for each row
for idx, row in df.iterrows():
    utt_id = str(row['utt_id']).strip()
    text = str(row['normalized_text']).strip()

    if not text or pd.isna(text):
        print(f"Skipping {utt_id} due to empty text.")
        continue

    output_path = f"output_wavs/{utt_id}.wav"
    try:
        tts.tts_to_file(text=text, file_path=output_path)
        print(f" Generated: {output_path}")
    except Exception as e:
        print(f" Failed for {utt_id}: {e}")
```
